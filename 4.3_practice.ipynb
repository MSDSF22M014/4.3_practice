```python
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([1.4, 1.6, 2.5, 2.6, 3.5, 3.7, 4.0])
N = 7 
xbar = sum(x)/N
ybar = sum(y)/N

b1 = (sum((x - xbar)*(y - ybar)))/(sum((x - xbar)**2))
b0 = ybar - b1*xbar
print("Slope: ", b1)
print("Y-intercept: ", b0)
```

```python
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([1.4, 1.6, 2.5, 2.6, 3.5, 3.7, 4.0])
cov_mat = np.cov(x, y)
print("Covariance matrix: \n",cov_mat)

b1 = cov_mat[0,1]/cov_mat[0,0]       
b0 = sum(y)/N  - b1 * sum(x)/N

print("\nSlope: ", b1)
print("Y-intercept: ", b0)
```
